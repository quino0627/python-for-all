```python
class A:
    a=3
    b=4
    @classmethod #decorator
    def sun(cls): #class method!
        return 'sun'
    @staticmethod
    def star(): #기본적인 인자가 필요 없음. instance에서도 쓸 수 있고, class에서도 쓸 수 있음.
        return 'star'
    def moon(self,t): #insance method
        self.t=t#instance attribute/variable
    def __init__(self,x=0): #instance 하는 순간 실행되는 method
        self.x=x
        print('초기화')
```
