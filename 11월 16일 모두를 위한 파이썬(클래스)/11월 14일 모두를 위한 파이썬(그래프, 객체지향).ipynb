```python
def yyy(**y):
    return y
#**y : keyword 방식이다 이말이야! 
```

```python
yyy() 
#인자를 전달하지 않아도 함수호출이 된다 이말이야.
```

```python
yyy(a=1,b=2,c=3)
#인자를 전달하면 dict 형태로 값을 반환한다 이말이야
```
